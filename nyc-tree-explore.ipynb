```python
tree_problems = tree_census_subset[['root_stone',
       'root_grate', 'root_other', 'trunk_wire', 'trnk_light', 'trnk_other',
       'brch_light', 'brch_shoe', 'brch_other']]

# Get actual counts
tree_problems_counts = tree_problems.apply(pd.Series.value_counts)

# Get percentages
tree_problems_percent = tree_problems.apply(lambda x: x.value_counts(normalize=True) * 100)

# Combine counts and percentages
tree_problems_combined = tree_problems_counts.apply(lambda x: x.astype(str) + ' (' + tree_problems_percent[x.name].round(2).astype(str) + '%)')

tree_problems_combined
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you've already created tree_problems_counts and tree_problems_percent as before
tree_problems_counts = tree_problems.apply(pd.Series.value_counts)
tree_problems_percent = tree_problems.apply(lambda x: x.value_counts(normalize=True) * 100)

# Prepare data for plotting
categories = tree_problems_counts.columns
yes_counts = tree_problems_counts.loc['Yes']
no_counts = tree_problems_counts.loc['No']
yes_percentages = tree_problems_percent.loc['Yes']

# Create the stacked bar chart
fig, ax1 = plt.subplots(figsize=(15, 8))

# Plot the stacked bars
ax1.bar(categories, no_counts, label='No')
ax1.bar(categories, yes_counts, bottom=no_counts, label='Yes')

# Customize the primary y-axis (counts)
ax1.set_ylabel('Count')
ax1.tick_params(axis='y')

# Create a secondary y-axis for percentages
ax2 = ax1.twinx()
ax2.set_ylabel('Percentage')
ax2.set_ylim(0, 100)

# Add percentage labels on top of the 'Yes' bars
for i, (count, percentage) in enumerate(zip(yes_counts, yes_percentages)):
    ax1.text(i, no_counts[i] + count, f'{count}\n({percentage:.1f}%)', 
             ha='center', va='bottom')

# Set the title and labels
plt.title('Tree Problems: Counts and Percentages', fontsize=16)
ax1.set_xlabel('Problem Categories')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend
ax1.legend(loc='upper left')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
